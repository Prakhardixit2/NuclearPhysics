```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import pandas as pd
import statistics
from scipy.stats import norm

df = pd.read_excel(r'C:\Users\ACER\Downloads\hall.xlsx')
print(df)
```

```python
V=[]
for i in df["corrected voltage"]:
    V.append(i)
print(V)

T=[]
for i in df["temperature"]:
   T.append(i) 
print(T)
```

```python
def RH(V):
    t=0.5
    I=4
    H=3253
    Rh=(V*t)*pow(10,4)/(I*H)
    return Rh

Rh=[]
for i in range(len(V)):
    Rh.append(RH(V[i]))
    
print(Rh)
                       
```

```python
plt.xlabel("Temperature $^{o}C$")
plt.ylabel("Hall Coefficient $10^{3}*cm^{3}*C^{-1}$")
plt.grid("TRUE")
plt.suptitle(r"Hall Coefficient vs. Temperature")
plt.plot(np.array(T),np.array(Rh),"r*")
plt.plot(np.array(T),np.array(Rh))

x=np.array(T)
y=np.array(Rh)

for xy in zip(x, y):
   plt.annotate('(%.2f, %.2f)' % xy, xy=xy)
```
